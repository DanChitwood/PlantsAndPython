```python
# If the code takes too long to execute, consider reducing the number of times the loop iterates

import random
import matplotlib.pyplot as plt

X = [0]
Y = [0]
for n in range(100000):
    r = random.uniform(0, 100)
    if r < 1.0:
        x = 0
        y = 0.16*Y[n-1]
    elif r < 86.0:
        x = 0.85*X[n-1] + 0.04*Y[n-1]
        y = -0.04*X[n-1] + 0.85*Y[n-1]+1.6
    elif r < 93.0:
        x = 0.2*X[n-1] - 0.26*Y[n-1]
        y = 0.23*X[n-1] + 0.22*Y[n-1] + 1.6
    else:
        x = -0.15*X[n-1] + 0.28*Y[n-1]
        y = 0.26*X[n-1] + 0.24*Y[n-1] + 0.44
    X.append(x);Y.append(y)

plt.figure(figsize = [15,15])
plt.scatter(X,Y,color = 'k',alpha=0.1, s=2)
plt.axes().set_aspect('equal', 'datalim')
plt.axis('off')
```

## Genetic knock-outs

One of the go-to approaches of biologists is genetics. A fundamental feature of life is inheritance: progeny ***inherit*** traits from their parents. Overall, the progeny may be similar or substantially different from their parents and siblings, but offspring owe their phenotype (what the organism ***is***, a composite of its traits) to the inheritance of genetic material material and the effects of their environment. 

There is a biochemical basis to inheritance: DeoxyriboNucleic Acid (DNA). DNA replicates and is inherited by daughter cells that divide and across genetic lineages from parents to their offspring. DNA can create temporary copies of itself called RiboNucleic Acid (RNA) through a process called ***transcription***. The components of the RNA can then be ***translated*** into proteins. Both DNA and RNA are polymers composed of nucleotides, and proteins are polymers of amino acids. In the process of translation, the identity of RNA nucleotides is interpreted into the string of amino acids that compose a protein. Proteins, at a biochemical and cellular level, are the structural and enzymatic workhorses of the cell. The structure of proteins, determined by their amino acid sequence (ultimately derived from a genetic, nucleotide sequence) dictates the chemical conversions they undertake, their structural roles within a cell, or their critical roles in regulating the activity of other genes and ultimately the development of an organism and its morphology.

If the activity of a gene (a sequence of nucleotides that produce an RNA or protein product, or the existence of which is linked to an outcome) is completely abolished, either through a mutation (a change in its nucleotide sequence) or even whole parts of it are deleted, it's called a knock-out, or a null mutation. Knock-outs are a vital tool for biologists to determine what a gene "does", what the "function" of the gene is.

In the radio example, a genetic knock-out is like taking a part of the radio out. Does the radio still work? Is the function of the radio impaired or different? The strategy is the same in biology.

## Gene expression

Knowing what a gene ***does*** is very informative. But there is other information that is useful as well. For example, where does a gene ***act***? Just before the ***coding sequence*** of a gene (the part that encodes the nucleotides corresponding to amino acids) are nucleotides in the DNA called a ***promoter***. Proteins called ***transcription factors*** bind nucleotides to activate transcription, the process that makes RNA that either functions itself or is translated into proteins. During development, the process by which cells ***differentiate*** into different tissues and organs, promoters can activate the transcription of different genes at different times and in different places. The spatio-temporal ***expression*** of genes is one of the principal mechanisms by which different cells, tissues, and organs are created, a mechanism by which multicellular organisms ***develop***.

There are many ways to find out where a gene is expressed. You can use a nucleic acid probe that is ***complementary*** (the complementarity of double-stranded nucleic acids allows them to replicate) to the target messenger RNA (mRNA) ***transcript*** that is chemically-modified with a label, that subsequently allows you to see in which cells the target transcripts accumulate. Within the genome, you can insert a ***reporter gene*** (for example, fluorescent proteins from bioluminescent organisms like jelly fish and fireflies) just in front of the gene's promoter. Wherever a promoter is active, it will then produce the reporter gene in those cells and mark them. You can all just insert a ***transgene*** into the genome, which contains the promoter plus a reporter gene, and it will insert into a random location and report where the gene is active.

There are many creative and clever ways that biologists have figured out to determine where a gene is active!
